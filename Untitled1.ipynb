```python
import pandas as pd
filename = "/sales_predictions_2023.csv"
df = pd.read_csv(filename)
df
num_filas, num_columnas = df.shape
print(f"1) Filas: {num_filas}, Columnas: {num_columnas}")

tipos_de_datos = df.dtypes
print("\n2) Tipos de datos por variable:")
print(tipos_de_datos)

df_sin_duplicados = df.drop_duplicates()
print("\n3) DataFrame sin duplicados:")
print(df_sin_duplicados)

valores_faltantes = df.isnull().sum()
print("\n4) Valores faltantes por variable:")
print(valores_faltantes)

df_sin_nulos = df.dropna()

valores_faltantes_despues = df_sin_nulos.isnull().sum()
print("\n6) Valores faltantes después de abordarlos:")
print(valores_faltantes_despues)

estadisticas_resumidas = df.describe()
print("\n8) Estadísticas resumidas para columnas numéricas:")
print(estadisticas_resumidas)
```
